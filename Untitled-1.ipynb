```python
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
```

```python
# Load the two datasets from Excel files
#df1 = pd.read_excel('ds1.xlsx')
df2 = pd.read_excel('ds2.xlsx')
```

```python
# Group the data by gender and time period
grouped_df = df2.groupby(['gender', 'timeperiod'])

# Loop through each group and access the data
for (gender, timeperiod), group_df in grouped_df:
    print(f"Gender: {gender}, Time Period: {timeperiod}")
    print(group_df[['readingscore', 'writingscore', 'mathscore', 'readingscoreSL', 'writingscoreSL']])
    print()
```

```python
# Calculate summary statistics for each group
summary_stats = grouped_df.agg({'readingscore': ['mean', 'median', 'std'],
                                 'writingscore': ['mean', 'median', 'std'],
                                 'mathscore': ['mean', 'median', 'std'],
                                 'readingscoreSL': ['mean', 'median', 'std'],
                                 'writingscoreSL': ['mean', 'median', 'std']})
print(summary_stats)
```

```python

# Create box plots for each score
plt.figure(figsize=(12, 8))
for i, score in enumerate(['readingscore', 'writingscore', 'mathscore', 'readingscoreSL', 'writingscoreSL']):
    plt.subplot(3, 2, i + 1)
    sns.boxplot(x='timeperiod', y=score, hue='gender', data=df2)
    plt.title(f'{score} Distribution by Time Period and Gender')
    plt.xlabel('Time Period')
    plt.ylabel(score)
    plt.legend(title='Gender', loc='upper right')   # 0 - female, 1-male
plt.tight_layout()
plt.show()


# Print summary statistics of the graph
for score in ['readingscore', 'writingscore', 'mathscore', 'readingscoreSL', 'writingscoreSL']:
    summary = df2.groupby(['timeperiod', 'gender'])[score].describe()
    summary['mean'] = df2.groupby(['timeperiod', 'gender'])[score].mean()
    print(f'{score} Summary Statistics:')
    print(summary[['25%', '50%', '75%', 'mean']])
    print()


```

```python

# Calculate and print combined mean for time periods (0, 1, 2) and (3, 4, 5) for each gender
for score in ['readingscore', 'writingscore', 'mathscore', 'readingscoreSL', 'writingscoreSL']:
    combined_mean = df2[df2['timeperiod'].isin([0, 1, 2])].groupby('gender')[score].mean()
    print(f'Combined Mean of Time Periods (0, 1, 2) for {score}:')
    print(combined_mean)
    combined_mean = df2[df2['timeperiod'].isin([3, 4, 5])].groupby('gender')[score].mean()
    print(f'Combined Mean of Time Periods (3, 4, 5) for {score}:')
    print(combined_mean)
    print()

# plotting graphs of above statistics
scores = ['readingscore', 'writingscore', 'mathscore', 'readingscoreSL', 'writingscoreSL']
for score in scores:
    combined_mean_0_2 = df2[df2['timeperiod'].isin([0, 1, 2])].groupby('gender')[score].mean()
    combined_mean_3_5 = df2[df2['timeperiod'].isin([3, 4, 5])].groupby('gender')[score].mean()

    # Plotting
    fig, ax = plt.subplots(figsize=(8, 6))
    x = range(len(combined_mean_0_2))
    width = 0.35

    bars1 = ax.bar(x, combined_mean_0_2, width, label='before covid')
    bars2 = ax.bar([p + width for p in x], combined_mean_3_5, width, label='after covid')

    ax.set_xlabel('Gender')
    ax.set_ylabel('Mean')
    ax.set_title(f'Combined Mean of Time Periods for {score}')
    ax.set_xticks([p + width / 2 for p in x])
    ax.set_xticklabels(['Female', 'Male'])
    ax.legend()

    plt.show()
```

```python
# violin plot
plt.figure(figsize=(12, 8))
for i, score in enumerate(['readingscore', 'writingscore', 'mathscore', 'readingscoreSL', 'writingscoreSL']):
    plt.subplot(3, 2, i + 1)
    sns.violinplot(x='timeperiod', y=score, hue='gender', data=df2, split=True)
    plt.title(f'{score} Distribution by Time Period and Gender')
    plt.xlabel('Time Period')
    plt.ylabel(score)
    plt.legend(title='Gender', loc='upper right')
plt.tight_layout()
plt.show()
```
