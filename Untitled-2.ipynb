```python
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
```

```python
# read data
df = pd.read_excel('ds1.xlsx')
```

```python
# Assuming 'Gender' is the name of your gender column
gender_counts = df['Gender'].value_counts()

# Display the counts
print(gender_counts)
```

```python
# Group by gender and the digital tool column
# 1. Before Covid 19 which digital tools do you use
grouped_data_before = df.groupby(['Gender', 'Before COVID-19: Which of the following digital tools do you usually use?']).size().unstack(fill_value=0)

# Display the grouped data
print(grouped_data_before)
```

```python
# Grouped bar plot
grouped_data_before.plot(kind='bar')
plt.title('Digital Tools Used Before COVID-19 by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Legend', loc='upper right')
plt.show()
```

```python
# Group by gender and the digital tool column
# After Covid 19 which digital tools do you use
grouped_data_after = df.groupby(['Gender', 'After COVID-19: Which of the following digital tools do you usually use?']).size().unstack(fill_value=0)

# Display the grouped data
print(grouped_data_after)
```

```python
# Grouped bar plot
grouped_data_after.plot(kind='bar')
plt.title('Digital Tools Used After COVID-19 by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Legend', loc='upper right')
plt.show()
```

```python
# Reorder the columns so that "Before" and "After" for each tool are next to each other
combined_data = pd.concat([grouped_data_before.add_suffix(' (Before)'), grouped_data_after.add_suffix(' (After)')], axis=1)
ordered_columns = []

# Extract the unique digital tools (without the "Before" and "After" suffixes)
digital_tools_agree = [col.replace(' (Before)', '').replace(' (After)', '') for col in combined_data.columns]
digital_tools_agree = sorted(set(digital_tools_agree))

# Create the order of columns with each tool's "Before" and "After" next to each other
for digital_tool_agree in digital_tools_agree:
    ordered_columns.append(digital_tool_agree + ' (Before)')
    ordered_columns.append(digital_tool_agree + ' (After)')

# Reindex the DataFrame with the ordered columns
combined_data = combined_data[ordered_columns]

# Normalize the data by converting counts to percentages
normalized_data = combined_data.div(combined_data.sum(axis=1), axis=0) * 100

print(combined_data)
print(normalized_data)

#print(normalized_data.to_markdown())
```

```python
# Grouped bar plot with reordered columns
ax = combined_data.plot(kind='bar', color=['#1f77b4', '#1f77b4', '#ff7f0e', '#ff7f0e', '#2ca02c', '#2ca02c', '#d62728', '#d62728', '#9467bd', '#9467bd'])
plt.title('Digital Tools Used Before and After COVID-19 by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Digital Tools', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()


ax = normalized_data.plot(kind='bar', color=['#1f77b4', '#1f77b4', '#ff7f0e', '#ff7f0e', '#2ca02c', '#2ca02c', '#d62728', '#d62728', '#9467bd', '#9467bd'])
plt.title('Percentage of Digital Tools Used Before and After COVID-19 by Gender')
plt.xlabel('Gender')
plt.ylabel('Precentage')
plt.legend(title='Digital Tools', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
